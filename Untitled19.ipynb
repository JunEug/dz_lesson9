```python
import pandas as pd
```

```python
df = pd.read_csv("./dz_files/dz9/water.csv", delimiter=",")
```

```python
print(df)
```

```python
import seaborn as sns

sns.scatterplot(x="hardness", y="mortality", data=df)
```

```python
from scipy.stats import pearsonr, spearmanr

pearson_corr, _ = pearsonr(df["hardness"], df["mortality"])
spearman_corr, _ = spearmanr(df["hardness"], df["mortality"])

print(f"Pearson correlation coefficient: {pearson_corr}")
print(f"Spearman correlation coefficient: {spearman_corr}")
```

```python
from scipy.stats import linregress

slope, intercept, _, _, _ = linregress(df["hardness"], df["mortality"])

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

```python
from sklearn.metrics import r2_score

predicted_mortality = slope * df["hardness"] + intercept
r2 = r2_score(df["mortality"], predicted_mortality)

print(f"R-squared: {r2}")
```

```python
copyresiduals = df["mortality"] - predicted_mortality

sns.scatterplot(x=df["hardness"], y=copyresiduals)
```
